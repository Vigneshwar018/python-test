```python
import numpy as np 
import pandas as pd 
import datetime

fl = pd.read_csv('icc/india-player.csv')
```

```python
fl.head()
```

```python
#fl['ODI Debut'] = pd.to_datetime(fl['ODI Debut']) 
fl['Date of birth'] = pd.to_datetime(fl['Date of birth']) 
```

```python
deb = fl['Debut Team'].apply(lambda x: x.split(' v'))
```

```python
fl['Team'] = fl['Team'].str.capitalize()

def te(x):
    if x[0] == fl['Team'].all():
        return x[1]
    else:
        return x[0]
        
fl['Debut Team'] = deb.apply(te)
```
