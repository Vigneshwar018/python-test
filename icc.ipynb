```python
import numpy as np 
import pandas as pd 
import datetime

fl = pd.read_csv('icc/india-player.csv')
```

```python
fl['Date of birth'] = pd.to_datetime(fl['Date of birth'])
fl
```

```python
fl['Debut Team'] = fl['Debut Team'].str.lower()
deb = fl['Debut Team'].apply(lambda x: x.split(' v '))

```

```python
fl['Team'] = fl['Team'].str.lower()
k = iter(range(fl.shape[0]))
def te(x):
    global k
    if x[0] == (fl['Team'][next(k)]):
        return x[1]
    else:
        return x[0]
    
fl['Debut Team']= deb.apply(te)
#a = fl['Debut Team'][1962][0]
#b = fl['Team'][1962]
fl['Debut Team']
```

```python
pd.set_option('display.max_columns', 31)
pd.set_option('display.max_rows', 50)
```

```python
fl['Highe Score'] = fl['Highe Score'].str.replace('*','')
```

```python
def rep(x, a, b, c):
    if c == True:
        pass
    return x.replace(a,b)

fl = fl.apply(rep, args=('--','0', False ))
fl = fl.apply(rep, args=('NaN','0', False ))
```

```python
fl.fillna(value=0, inplace=True)
Dat = fl['ODI Debut'].str.replace(',', '-') 
fl['ODI Debut'] = pd.to_datetime(Dat)
```

```python
cols = fl.columns[7:-2]

fl[cols] = fl[cols].apply(pd.to_numeric)

#fl[fl['Bowling Style'] == 0]
fl['Bowling Style'] = fl['Bowling Style'].map(lambda s: str(s).replace('0', 'Not-Bowled')) 
fl['Batting Style'] = fl['Batting Style'].map(lambda s: str(s).replace('0', 'Right-hand bat')) 


#fl.to_csv('india_players.csv', index = False)
```
