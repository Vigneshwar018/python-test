```python
import numpy as np 
import pandas as pd 
import datetime

fl = pd.read_csv('icc/new zealand-player.csv')
```

```python
fl['Date of birth'] = pd.to_datetime(fl['Date of birth'])
fl
```

```python
fl['Debut Team'] = fl['Debut Team'].str.lower()
deb = fl['Debut Team'].apply(lambda x: x.split(' v'))
```

```python
fl['Team'] = fl['Team'].str.lower()

def te(x):
    if x[0] == fl['Team'].all():
        return x[1]
    else:
        return x[0]
        
fl['Debut Team'] = deb.apply(te)
```

```python
pd.set_option('display.max_columns', 31)
```

```python
fl['Highe Score'] = fl['Highe Score'].str.replace('*','')
```

```python
def rep(x, a, b):
    return x.replace(a,b)

fl = fl.apply(rep, args=('--','0'))
fl = fl.apply(rep, args=('NaN','0'))
```

```python
fl.fillna(value=0, inplace=True)
Dat = fl['ODI Debut'].str.replace(',', '-') 
fl['ODI Debut'] = pd.to_datetime(Dat)
```

```python
cols = fl.columns[7:-2]

fl[cols] = fl[cols].apply(pd.to_numeric)

fl[fl['Highe Score'] >= 180]
```
